## Analisis Acciones

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
simbolos = ["AAPL","AMZN","TSLA","FB","GOOG"]
url ="https://query1.finance.yahoo.com/v7/finance/download/{}?period1=1588123834&period2=1619659834&interval=1d&events=history&includeAdjustedClose=true"
```

```python
df = pd.DataFrame()
```

```python
lista_temp = list()
for simbolo in simbolos:
    df_temp = pd.read_csv(url.format(simbolo))
    df_temp["Symbol"] = simbolo
    lista_temp.append(df_temp)
```

```python
df = pd.concat(lista_temp,ignore_index=True)
```

```python
df.describe()
```

```python
df["Volat"] =( df["High"] - df["Low"] ) / df["Open"]
```

```python
df["Return"] = (df["Close"] - df["Open"]) - 1.0
```
